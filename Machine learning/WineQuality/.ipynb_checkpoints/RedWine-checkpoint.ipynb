## Ploting the all columns with alcohol in redwine dataframe

```python
plt.figure(figsize=(50,40))
plt.subplot(3,4,1)
plt.scatter(red_wine['fixed acidity'],red_wine.alcohol)
plt.title('fixed acidity')
plt.subplot(3,4,2)
plt.scatter(red_wine['volatile acidity'],red_wine.alcohol)
plt.title('volatile acidity')

plt.subplot(3,4,3)
plt.scatter(red_wine['citric acid'],red_wine.alcohol)
plt.title('citric acid')

plt.subplot(3,4,4)
plt.scatter(red_wine['residual sugar'],red_wine.alcohol)
plt.title('residual sugar')

plt.subplot(3,4,5)
plt.scatter(red_wine['chlorides'],red_wine.alcohol)
plt.title('chlorides')

plt.subplot(3,4,6)
plt.scatter(red_wine['free sulfur dioxide'],red_wine.alcohol)
plt.title('free sulfur dioxide')

plt.subplot(3,4,7)
plt.scatter(red_wine['total sulfur dioxide'],red_wine.alcohol)
plt.title('total sulfur dioxide')

plt.subplot(3,4,8)
plt.scatter(red_wine['density'],red_wine.alcohol)
plt.title('density')

plt.subplot(3,4,9)
plt.scatter(red_wine['pH'],red_wine.alcohol)
plt.title('pH')

plt.subplot(3,4,10)
plt.scatter(red_wine['sulphates'],red_wine.alcohol)
plt.title('sulphates')

plt.subplot(3,4,11)
plt.scatter(red_wine['quality'],red_wine.alcohol)
plt.title('quality')
```

## Distplot

```python
plt.figure(figsize=(50,40))
plt.subplot(3,4,1)
sns.distplot(red_wine['fixed acidity'])
plt.subplot(3,4,2)
sns.distplot(red_wine['volatile acidity'])
plt.subplot(3,4,3)
sns.distplot(red_wine['citric acid'])
plt.subplot(3,4,4)
sns.distplot(red_wine['residual sugar'])
plt.subplot(3,4,5)
sns.distplot(red_wine['chlorides'])
plt.subplot(3,4,6)
sns.distplot(red_wine['free sulfur dioxide'])
plt.subplot(3,4,7)
sns.distplot(red_wine['total sulfur dioxide'])
plt.subplot(3,4,8)
sns.distplot(red_wine['density'])
plt.subplot(3,4,9)
sns.distplot(red_wine['pH'])
plt.subplot(3,4,10)
sns.distplot(red_wine['sulphates'])
plt.subplot(3,4,11)
sns.distplot(red_wine.alcohol)
plt.subplot(3,4,12)
sns.distplot(red_wine['quality'])
```
