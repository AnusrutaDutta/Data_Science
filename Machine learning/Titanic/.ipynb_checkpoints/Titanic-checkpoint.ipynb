```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
titanic.Age=titanic['Age'].fillna(round(titanic.Age.mean()))
```

```python
titanic.isnull().sum()
```

```python
titanic=titanic.drop(['PassengerId','Name','Ticket','Cabin'],1)
titanic.head()
```

```python
dummy=pd.get_dummies(titanic.Sex,drop_first=False)
dummy.head()
```

```python
titanic=pd.concat([titanic,dummy],axis=1)
```

```python
titanic.head()
```

```python
titanic=titanic.drop(['Sex'],1)
```

```python
titanic.head()
```

```python
titanic.groupby('Embarked').count()
```

```python
dummy_2=pd.get_dummies(titanic.Embarked,drop_first=False)
dummy_2
```

```python
titanic=pd.concat([titanic,dummy_2],axis=1)
```

```python
titanic=titanic.drop(['Embarked'],1)
```
