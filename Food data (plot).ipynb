# Sum of total food consumption for each country

```python
Ace=Foods.sum(axis=1)
Ace
```

# Sum of total food consumption

```python
Ace.sum()
```

# Creating the new column of % of the total food consumption

```python
Luffy=[round(100*i/Ace.sum(),2) for i in Ace]
Luffy
```

```python
Foods['%_of_Foods_consumed_countrywise']=Luffy
Foods
```

```python
Foods['%_of_Foods_consumed_countrywise'].mean()
```

```python
np.median(Foods['%_of_Foods_consumed_countrywise'])
```

```python
stats.mode(Foods['%_of_Foods_consumed_countrywise'])
```

```python
Foods['%_of_Foods_consumed_countrywise'].kurtosis()
```

```python
Foods['%_of_Foods_consumed_countrywise'].skew()
```

```python
sns.boxplot(Foods['%_of_Foods_consumed_countrywise'])
```

```python
Quan1=Foods['%_of_Foods_consumed_countrywise'].quantile(0.25)
```

```python
Quan3=Foods['%_of_Foods_consumed_countrywise'].quantile(0.75)
IQR= Quan3 - Quan1
IQR
```

# Ploting the graph of percentage

```python
plt.plot(Foods['%_of_Foods_consumed_countrywise'],lw=5,ls='--',marker='o',markersize=10,markerfacecolor='magenta',markeredgecolor='red')
```

# Sort by % of total food consumption

```python
food_sort=Foods.sort_values(by=['%_of_Foods_consumed_countrywise'])
food_sort
```
